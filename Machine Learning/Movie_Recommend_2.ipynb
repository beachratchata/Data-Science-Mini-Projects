## Movie Recommendation System : ระบบแนะนำหนัง

## Collaborative Filtering 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#### df1 : ตารางชื่อหนัง

```python
df1 = pd.read_csv('title.txt')
df1.head()
```

#### df2 : ตารางการให้คำแนนหนัง 

```python
df2 = pd.read_csv('rating.txt' , sep='\t' , names=['user_id' , 'item_id' , 'rating' , 'timestamp'])
df2.head()
```

#### นำ 2 ตารางที่ movie id เหมือนกันมารวมเข้าด้วยกัน

```python
movie = pd.merge(df1 , df2 , on='item_id' )
movie.drop(['timestamp'] , axis=1 , inplace=True)
movie.sample(5)
```

```python
movie['title'] = movie['title'].str.replace('\(.*\)','').str.strip()
```

```python
movie.sample(5)
```

#### ทำเป็นตาราง pivot 

```python
movie_rating = pd.pivot_table(movie , index='user_id' , columns='title' , values='rating')
```

```python
movie_rating.head()
```

#### หนังที่มีผู้โหวดไม่ถึง 10 คน ถูกลบออกไป

```python
movie_rating.dropna(thresh=10 , axis=1 , inplace=True)
```

```python
movie_rating.shape
```

#### แทนค่าว่างด้วย 0

```python
movie_rating.fillna(0 , inplace=True)
```

```python
movie_rating.sample(5)
```

#### หาความสัมพันธ์กันของหนังด้วยตาราง correlation 

```python
item_corr = movie_rating.corr()
```

```python
item_corr.head()
```

```python
#### 
```

```python
def similar_movie(movie , rating) :
    similar_score = item_corr[movie] * (rating - 2.5)
    #similar_score = similar_score.sort_values(ascending=False)
    return similar_score
```

#### 10 อันดับหนังแนะนำ เมื่อให้คะแนนหนังเรื่อง Batman 5 คะแนน

```python
batman_5 = similar_movie('Batman' , 5)
batman_5.sort_values(ascending=False)[1:11]
```

#### 10 อันดับหนังแนะนำ เมื่อให้คะแนนหนังเรื่อง Batman 1 คะแนน

```python
batman_1 = similar_movie('Batman' , 1)
batman_1.sort_values(ascending=False)[:10]
```

```python
my_rating = pd.DataFrame([('Back to the Future' , 3) , 
                            ('Star Wars' , 4 ),
                           ('Aliens' , 2)] , columns=['movie','rating'])
```

#### ให้คะแนนหนัง 3 เรื่อง : Back to the Future = 3 , Star Wars = 4 , Aliens =  2

```python
my_rating
```

```python
similar_df = pd.DataFrame()
```

```python
for movie_name , rating in zip(my_rating['movie'] , my_rating['rating']) :
    similar_df = similar_df.append(similar_movie(movie_name , rating))     
```

```python
similar_df
```

#### 10 อันดับหนังแนะนำเมื่อให้คะแนนหนังตาม my_rating

```python
similar_df.drop(my_rating['movie'].tolist() , axis=1).sum().sort_values(ascending=False)[:10]
```
