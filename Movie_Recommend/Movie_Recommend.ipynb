```python
import pandas as pd 
from sklearn.feature_extraction.text import CountVectorizer , TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import json
```

```python
df1 = pd.read_csv('tmdb_5000_credits.csv.zip')
df2 = pd.read_csv('tmdb_5000_movies.csv.zip')
```

```python
df1.head()
```

```python
df2.head()
```

```python
movie = pd.merge(df1 , df2 , left_on='movie_id' , right_on='id')
```

```python
movie.head()
```

```python
movie.columns
```

```python
movie = movie[['title_x','cast', 'crew','genres','keywords','overview']].reset_index()
```

```python
movie.head()
```

```python
movie.rename(columns={'title_x' : 'title'} , inplace=True)
```

```python
movie.columns
```

```python
for i in movie[['cast', 'crew', 'genres', 'keywords']]:
    movie[i] = movie[i].apply(json.loads)
```

```python
movie.head()
```

```python
def lis_name(x):
    lis = []
    for i in x :
        lis.append(i['name'])     
    return lis
```

```python
def director_name(x):
    lis = []
    for i in x :
        if i['job'] == 'Director':
            lis.append(i['name'])
    return lis
```

```python
for i in movie[['cast','genres','keywords']]:
    movie[i] = movie[i].apply(lis_name)
```

```python
movie['crew'] = movie['crew'].apply(director_name)
```

```python
movie.head()
```

```python
movie.isna().sum()
```

```python
movie['overview'] = movie['overview'].fillna('') 
```

```python
movie['features']  = movie['cast'] + movie['crew'] + movie['genres'] + movie['keywords']
```

```python
movie['features'] = movie['features'].apply(lambda x : ' '.join(x))
```

```python
movie['features'] = movie['features'] + movie['overview']
```

```python
tv = TfidfVectorizer(stop_words='english')
```

```python
def movie_reccomend(movie_name : str , top : int):
    count_metrix = tv.fit_transform(movie['features'])
    cosin_sim = cosine_similarity(count_metrix)
    
    movie_index = movie[movie['title'] == movie_name].index.values[0]
    similar_movie = list(enumerate(cosin_sim[movie_index]))
    
    sort_similar_movie = sorted(similar_movie , key=lambda x : x[1] , reverse=True)
    
    for i , s in sort_similar_movie[1:top]:
        print(movie['title'][i])
```

```python
movie_reccomend('Batman' , 10)
```

```python
movie_reccomend('X-Men' , 10)
```

```python
movie_reccomend('Iron Man' , 10)
```
